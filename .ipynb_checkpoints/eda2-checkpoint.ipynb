```python
labels = pd.read_csv("train_labels.csv")
labels
```

```python
labels["question_number"] = labels["session_id"].apply(lambda x: x.split("_")[1][1:])
labels["session_id"] = labels["session_id"].apply(lambda x: x.split("_")[0])
labels["question_number"] = labels["question_number"].apply(int)
labels["session_id"] = labels["session_id"].apply(int)
labels.sort_values("session_id", inplace = True)
labels.set_index("session_id", inplace= True)
```

```python
labels=labels[["question_number", "correct"]]
labels
```

```python
correct_ratios = []
for q in range(1, 19):
    tmp = labels[labels["question_number"] == q]['correct']
    ratio = tmp.sum() / len(tmp) 
    correct_ratios.append(ratio)

xrange = range(1, 19)
mean = np.mean(correct_ratios)
plt.figure(figsize=(8, 4))
plt.bar(x=xrange, height=correct_ratios, color='purple')
plt.axhline(mean, color='red', ls='--', lw=3)
plt.xticks(xrange)
plt.legend([f'Mean: {mean:.3f}'])
plt.title("Correct answers per question")
plt.xlabel("Question number")
plt.ylabel("Ratio")
plt.show()
```

```python
EVENTS = ['navigate_click','person_click','cutscene_click','object_click','map_hover','notification_click','map_click','observation_click',
          'checkpoint']
CATS = ['event_name', 'fqid', 'room_fqid']
NUMS = ['elapsed_time','level','room_coor_x', 'room_coor_y', 'screen_coor_x', 'screen_coor_y']
```

```python
dfs = [] # empty dataset 
    
    #getting number of unique events
for c in CATS:
    tmp = train.groupby(['session_id','level_group'])[c].agg('nunique')
    tmp.name = tmp.name + '_nunique'
    dfs.append(tmp)
    

for c in NUMS:
    tmp = train.groupby(['session_id','level_group'])[c].agg('mean')
    tmp.name = tmp.name + '_mean'
    dfs.append(tmp)
for c in NUMS:
    tmp = train.groupby(['session_id','level_group'])[c].agg('std')
    tmp.name = tmp.name + '_std'
    dfs.append(tmp)
for c in EVENTS: 
    train[c] = (train.event_name == c).astype('int8')
for c in EVENTS + ['elapsed_time']:
    tmp = train.groupby(['session_id','level_group'])[c].agg('sum')
    tmp.name = tmp.name + '_sum'
    dfs.append(tmp)
temp = train.drop(EVENTS,axis=1)
        
temp = pd.concat(dfs,axis=1)
temp = temp.fillna(-1)
temp = temp.reset_index()
temp = temp.set_index('session_id')
temp
```

```python
ana = []
for t in range(1,19): #each question number
        
    # USE THIS TRAIN DATA WITH THESE QUESTIONS
    if t<=3: grp = '0-4'
    elif t<=13: grp = '5-12'
    elif t<=22: grp = '13-22'

    ana.append(pd.merge(temp[temp["level_group"]== grp], 
             labels[labels["question_number"]== t],
             on = "session_id", how = "left"))
ana = pd.concat(ana, axis = 0)
ana
```

```python
ana.corr()
plt.figure(figsize=(25, 25))
mask = np.triu(np.ones_like(ana.corr(), dtype=bool))
heatmap = sns.heatmap(abs(ana.corr()), mask = None, vmin=0, vmax=1, annot=True, cmap= "RdBu")
heatmap.set_title('Correlation map', fontdict={'fontsize':18}, pad=18);
```

```python
for grp in train["level_group"].unique().tolist():
    plot = ana[ana["level_group"] == grp].corr()
    plt.figure(figsize=(25, 25))

    # Store heatmap object in a variable to easily access it when you want to include more features (such as title).
    # Set the range of values to be displayed on the colormap from 0 to 1, and set the annotation to True to display the correlation values on the heatmap.
    heatmap = sns.heatmap(abs(plot.corr()), mask = None, vmin=0, vmax=1, annot=True, cmap= "RdBu")
    # Give a title to the heatmap. Pad defines the distance of the title from the top of the heatmap.
    heatmap.set_title('Correlation Heatmap: {}'.format(grp), fontdict={'fontsize':18}, pad=18);
```

```python
FEATURES = [c for c in temp.columns if c != 'level_group']
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

train_model = {}
for t in range(1,19): #each question number
        
    # USE THIS TRAIN DATA WITH THESE QUESTIONS
    if t<=3: grp = '0-4'
    elif t<=13: grp = '5-12'
    elif t<=22: grp = '13-22'
        
    X = temp[temp["level_group"] == grp][FEATURES].sort_index()
    y = labels[labels['question_number'] == t]["correct"]

    # train and validation set split
    X_train, X_val, y_train, y_val = train_test_split( X, y, test_size=0.20, random_state=42)

    clf = make_pipeline(StandardScaler(),
                        LogisticRegression(random_state=0, max_iter=10000, verbose = 0))
    
    clf.fit(X_train, y_train)
    
    print("Question number {}".format(t))
    print("Model Accuracy: {}".format(metrics.accuracy_score(clf.predict(X_val), y_val)))
    print("The f1 score of the model is {}".format(metrics.f1_score(clf.predict(X_val), y_val)))
    # SAVE MODEL, PREDICT VALID OOF
    train_model[f'{grp}_{t}'] = clf
    train_model
```
